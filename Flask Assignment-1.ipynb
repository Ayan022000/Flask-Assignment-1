# Q 1.

## What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight, open-source Python web framework that allows developers to build web applications quickly and easily. It is based on the WSGI (Web Server Gateway Interface) toolkit and Jinja2 template engine. Flask is known for its simplicity and flexibility, which makes it a popular choice for small to medium-sized web applications.

Some advantages of using Flask framework are:

Lightweight: Flask is a lightweight framework, which means it has a minimalistic core that makes it easy to learn and use.

Flexibility: Flask is highly flexible and can be easily extended to add new features and functionalities as per the requirements.

Easy to learn: Flask follows a simple and intuitive design, which makes it easy to learn even for beginners.

Compatibility: Flask works well with a wide range of third-party extensions and libraries, making it a versatile choice for web development.

Built-in Development Server: Flask comes with a built-in development server that makes it easy to test and debug the application during the development phase.

## What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs to specific functions or views that handle those URLs. This is done using the @app.route() decorator, which is used to define the URL endpoint and the HTTP methods that the endpoint will accept.

## What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

The url_for() function is used in Flask for URL Building. It generates a URL to a specific endpoint based on the name of the function that handles that endpoint.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/about')
def about():
    return 'About us'

if __name__ == '__main__'
    app.run()
```
